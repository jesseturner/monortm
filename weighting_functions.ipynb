Visualization of the weighting functions generated by the monoRTM line-by-line model. 

The information for running the RTM is in my project notes. 

```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
monortm_07_ds = xr.open_dataset('MONORTM_07.nc')
monortm_14_ds = xr.open_dataset('MONORTM_14.nc')
```

```python
#--- Checking to make sure I have correct wavelengths

c = 3e8 #m s-1

wl_07 = c/(monortm_07_ds.FREQUENCY.values[0]*(1e9))
wl_14 = c/(monortm_14_ds.FREQUENCY.values[0]*(1e9))

print("wl band 07:", round(wl_07*1e6, 1), "um")
print("wl band 14:", round(wl_14*1e6, 1), "um")

```

```python
monortm_07_ds.LAYERS
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(monortm_07_ds.LAYER_OPTICAL_DEPTH, monortm_07_ds.LAYERS, label="Band07 (3.9um)", color="#3772FF")
ax.plot(monortm_14_ds.LAYER_OPTICAL_DEPTH, monortm_14_ds.LAYERS, label="Band14 (11.2um)", color="#DF2935")

ax.set_title("LBL weighting functions")
ax.set_xlabel("Weighting function")
ax.set_ylabel("Vertical layers")
ax.legend()

fig.show()
```
